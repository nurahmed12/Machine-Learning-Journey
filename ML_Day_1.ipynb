Standard deviation (SD) — one number for "how spread out"

Idea: SD tells how far values are from the average (mean), on average.

Steps to get SD (population):

Mean = sum of numbers ÷ count (n).

For each number, find distance = number − mean.

Square each distance (makes them positive).

Sum the squared distances.

Variance = sum of squares ÷ n.

SD = square root of variance.

Quick formula: SD = sqrt( (Σ (x - mean)²) / n )

If you only have a sample (not all data): divide by (n−1) instead of n (small technical fix).

Tiny intuition: Squaring prevents negative and positive distances canceling. Averaging gives one number. Square root returns to original units.

Percentile — where you sit compared to others

Idea: The p-th percentile is the value below which p% of data falls.

Simple position method (one common way):

Sort data from smallest to largest.

Compute position = p × (n + 1) where p is decimal (e.g., 40% → 0.40).

If position is whole (like 3), the percentile is the 3rd value. If not whole (like 2.4), go between the 2nd and 3rd values: move 0.4 of the gap.

Why (n + 1)? Think of data as fence posts. There are n posts but n+1 gaps (before first, between, after last). Using n+1 lets percentiles sit smoothly in gaps and treat 0% before first and 100% after last.

Note: There are other valid methods; (n+1) is simple and common.

Quartiles — percentiles that cut into 4 parts

Q1 = 25th percentile (25% below)

Q2 = 50th percentile = median (middle)

Q3 = 75th percentile (75% below)

IQR (interquartile range) = Q3 − Q1 (shows spread of middle 50%)

How to find: use the percentile method above with p = 0.25, 0.50, 0.75.

Mini examples to remember

SD example (short): For numbers 1,5,22,6,32,11 → mean ≈ 12.83 → variance ≈ 117.14 → SD ≈ 10.82. (This means a typical value is ~10.8 away from the mean.)

Percentile example: For [10,20,30,40,50], 40th pct → position = 0.4×(5+1)=2.4 → 2nd=20, 3rd=30 → value = 20 + 0.4×(30−20) = 24.

One-line reminders you can memorize

SD = typical distance from the mean.

Percentile = value below which p% of data fall.

Quartiles = 25%, 50%, 75% cut points; IQR = middle spread.
